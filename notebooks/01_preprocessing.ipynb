```python
# Function for cleaning a dataframe
def clean_dataframe(df):
    # Drop duplicate rows
    df_cleaned = df.drop_duplicates()

    # Fill missing values with appropriate methods (mean, median, or a specific value)
    # Example: Fill numerical columns with the mean, categorical with 'Unknown' or 'NaN'
    df_cleaned = df_cleaned.fillna(df_cleaned.mean(numeric_only=True))  # Filling numeric with mean
    df_cleaned = df_cleaned.fillna('Unknown')  # Filling categorical with 'Unknown' (or you can use 'NaN')

    # Drop rows with any remaining missing values (optional)
    df_cleaned = df_cleaned.dropna(how='any')

    return df_cleaned
```

```python
# Affectiva sensors produce -1 or very large numbers when the sensor loses track. Replace these with NaN
import numpy as np
def clean_sensor_df(df, numeric_cols, invalid_values=[-1, -999, 9999]):
    df = df.copy()
    for c in numeric_cols:
        df[c] = pd.to_numeric(df[c], errors='coerce')  # force numeric
        df[c] = df[c].replace(invalid_values, np.nan)
    return df
```

```python
from tqdm import tqdm
```

```python
window_size = 5        # seconds
step_size   = 5        # non-overlapping windows; set <window_size for overlapping
time_col    = 'UnixTime'
student_col = 'Student_ID'
```

```python
# detect EEG band columns
eeg_band_cols = [c for c in df_eeg.columns if any(b in c.lower() for b in
                                                 ['alpha','beta','theta','delta','gamma'])]
print("EEG features:", eeg_band_cols[:10])
```

```python
# detect TIVA emotion columns
emo_cols = [c for c in df_tiva.columns if c.lower() in
            ('engagement','valence','joy','arousal','happiness','surprise','sadness','anger','disgust','fear','dominant_emotion')]
print("TIVA emotion/label columns:", emo_cols)
```

```python
# Choose label column (dominant emotion or valence discretized)
label_col = 'dominant_emotion' if 'dominant_emotion' in df_tiva.columns else 'Valence'
```

```python
# --- function to create windows ---
def create_windows(df_eeg, df_tiva, student_id, window_size, step_size, time_col, label_col):
    eeg_s = df_eeg[df_eeg[student_col]==student_id].sort_values(time_col)
    tiva_s = df_tiva[df_tiva[student_col]==student_id].sort_values(time_col)
    if eeg_s.empty or tiva_s.empty:
        return []
    start_time = max(eeg_s[time_col].min(), tiva_s[time_col].min())
    end_time   = min(eeg_s[time_col].max(), tiva_s[time_col].max())
    windows=[]
    t = start_time
    while t+window_size <= end_time:
        eeg_win  = eeg_s[(eeg_s[time_col]>=t) & (eeg_s[time_col]<t+window_size)]
        tiva_win = tiva_s[(tiva_s[time_col]>=t) & (tiva_s[time_col]<t+window_size)]
        if eeg_win.empty or tiva_win.empty:
            t += step_size
            continue
        # aggregate EEG band features
        eeg_feats={}
        for c in eeg_band_cols:
            eeg_feats[f"{c}_mean"]=eeg_win[c].mean()
            eeg_feats[f"{c}_std"]=eeg_win[c].std()
        # aggregate TIVA facial features (AU/emotions)
        tiva_feats={}
        for c in emo_cols:
            if c==label_col:
                continue
            tiva_feats[f"{c}_mean"]=tiva_win[c].mean()
            tiva_feats[f"{c}_std"]=tiva_win[c].std()
        # label: majority label or mean valence
        if label_col=='dominant_emotion':
            label = tiva_win[label_col].mode()[0] if not tiva_win[label_col].mode().empty else np.nan
        else:
            label = tiva_win[label_col].mean()  # valence numeric
        windows.append({
            'Student_ID': student_id,
            'start_time': t,
            'end_time': t+window_size,
            'label': label,
            **eeg_feats, **tiva_feats
        })
        t += step_size
    return windows
```

```python
#loop over all students and build dataset
all_windows=[]
for sid in tqdm(df_eeg[student_col].unique(), desc='Building windows'):
    all_windows.extend(create_windows(df_eeg, df_tiva, sid, window_size, step_size, time_col, label_col))
```

```python
df_windows = pd.DataFrame(all_windows)
print("Final window-level dataset:", df_windows.shape)
df_windows.to_csv("window_level_dataset.csv", index=False)
```

## Synchronize timestamps

```python
def ensure_unix_seconds(df, time_col='UnixTime'):
    df = df.copy()
    df[time_col] = pd.to_numeric(df[time_col], errors='coerce')
    if df[time_col].dropna().max() > 1e11:   # probably milliseconds
        df[time_col] = df[time_col] / 1000.0
    return df

df_eeg  = ensure_unix_seconds(df_eeg, 'UnixTime')
df_tiva = ensure_unix_seconds(df_tiva, 'UnixTime')
df_psy  = ensure_unix_seconds(df_psy, 'routineStart')
# ensure routineEnd too
df_psy['routineEnd'] = pd.to_numeric(df_psy['routineEnd'], errors='coerce')
if df_psy['routineEnd'].dropna().max() > 1e11:
    df_psy['routineEnd'] = df_psy['routineEnd'] / 1000.0
```

## Fast per-second resampling (aggregate to fixed sampling rate = 1 Hz)

This reduces size and aligns modalities by second.

```python
def resample_to_seconds(df, time_col='UnixTime', student_col='Student_ID', agg_funcs=['mean']):
    """
    Returns a DataFrame aggregated per student per integer-second timestamp.
    Numeric columns aggregated with mean/std.
    """
    df = df.copy()
    # floor seconds to integer
    df['sec'] = df[time_col].astype('float').floordiv(1).astype('int64')
    # keep numeric columns only (exclude time/student)
    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    numeric_cols = [c for c in numeric_cols if c not in [time_col,'sec',student_col]]
    # group by student + second and agg
    grouped = df.groupby([student_col, 'sec'])[numeric_cols].agg(['mean','std'])
    # flatten columns
    grouped.columns = [f"{col}_{stat}" for col, stat in grouped.columns]
    grouped = grouped.reset_index()
    return grouped
```

```python
# Example usage:
eeg_sec  = resample_to_seconds(df_eeg, 'UnixTime', 'Student_ID')
tiva_sec = resample_to_seconds(df_tiva, 'UnixTime', 'Student_ID')

print("EEG per-second shape:", eeg_sec.shape)
print("TIVA per-second shape:", tiva_sec.shape)
```

## Saving The Files Data.

```python
import os
def save_if_not_exists(df, filename):
    filepath = os.path.join(output_path, filename)
    if os.path.exists(filepath):
        print(f"⚠️ {filename} already exists — skipping.")
    else:
        df.to_csv(filepath, index=False)
        print(f"✅ Saved {filename}")
```
