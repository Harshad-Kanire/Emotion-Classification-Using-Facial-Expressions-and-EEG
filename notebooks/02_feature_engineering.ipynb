## EEG features
EEG already has band columns (e.g., Alpha_AF7, Beta_AF7, Theta_TP9)

```python
# detect band-like columns
def detect_band_cols(df):
    bands = ['delta','theta','alpha','beta','gamma']
    band_cols = {b: [c for c in df.columns if b in c.lower()] for b in bands}
    return band_cols

eeg_band_cols = detect_band_cols(df_eeg)
print({k: len(v) for k,v in eeg_band_cols.items()})
```

# Aggregate per trial

```python
def aggregate_eeg_bands_per_trial(df_psy, df_source, band_cols, time_col='UnixTime', student_col='Student_ID'):
    rows=[]
    # group df_source by student for speed
    source_by_student = {sid:g for sid,g in df_source.groupby(student_col)}
    for _, trial in tqdm(df_psy.iterrows(), total=len(df_psy), desc='EEG trial agg'):
        sid = trial[student_col]; s = trial['routineStart']; e = trial['routineEnd']
        seg = source_by_student.get(sid)
        if seg is None:
            rows.append({**trial.to_dict(), **{f'{b}_mean':np.nan for b in band_cols}})
            continue
        seg = seg[(seg[time_col]>=s) & (seg[time_col]<=e)]
        feat={}
        for band, cols in band_cols.items():
            if len(cols)>0 and not seg.empty:
                # mean across all band channels
                feat[f'{band}_mean'] = seg[cols].mean().mean()
                feat[f'{band}_std']  = seg[cols].stack().std()
            else:
                feat[f'{band}_mean'] = np.nan
                feat[f'{band}_std']  = np.nan
        # frontal asymmetry example: log(power_right) - log(power_left)
        if 'alpha_af7' in [c.lower() for c in seg.columns] and 'alpha_af8' in [c.lower() for c in seg.columns]:
            try:
                # adapt exact column names present in your data
                a7 = seg[[c for c in seg.columns if 'alpha_af7' in c.lower()]].mean().mean()
                a8 = seg[[c for c in seg.columns if 'alpha_af8' in c.lower()]].mean().mean()
                feat['alpha_asymmetry'] = np.log(a8+1e-9) - np.log(a7+1e-9)
            except:
                feat['alpha_asymmetry'] = np.nan
        rows.append({**trial.to_dict(), **feat})

    return pd.DataFrame(rows)

# if band columns exist, pass df_eeg (raw) to function
eeg_trial_feats = aggregate_eeg_bands_per_trial(df_psy, df_eeg, eeg_band_cols, time_col='UnixTime')
```

# bandpower per trial using Welch

```python
# function to compute bandpower via Welch
BANDS = {'delta':(1,4),'theta':(4,8),'alpha':(8,13),'beta':(13,30),'gamma':(30,50)}
from scipy.signal import welch

def bandpower_welch(sig, fs, band):
    fmin,fmax = band
    freqs, psd = welch(sig, fs=fs, nperseg=min(len(sig), fs*2))
    idx = np.logical_and(freqs >= fmin, freqs <= fmax)
    return np.trapz(psd[idx], freqs[idx]) if idx.any() else 0.0

def compute_bandpower_per_trial(df_eeg, df_psy, band_cols, fs=2.3, time_col='UnixTime', student_col='Student_ID'):
    rows=[]
    by_student = {sid:g for sid,g in df_eeg.groupby(student_col)}
    for _, trial in tqdm(df_psy.iterrows(), total=len(df_psy), desc='Bandpower per trial'):
        sid=trial[student_col]; s=trial['routineStart']; e=trial['routineEnd']
        seg = by_student.get(sid)
        if seg is None:
            rows.append({**trial.to_dict(), **{f'{b}_{ch}':np.nan for b,cols in band_cols.items() for ch in cols}})
            continue
        seg = seg[(seg[time_col]>=s) & (seg[time_col]<=e)]
        feat={}
        if seg.empty:
             for b, cols in band_cols.items():
                for ch in cols:
                     feat[f'{b}_{ch}'] = np.nan
        else:
            # extract signals per channel and band(assume signals as columns)
            for bname, brange in BANDS.items():
                 cols = band_cols.get(bname)
                 if cols:
                    for col in cols:
                        if col in seg.columns:
                            sig = seg[col].dropna().values
                            feat[f'{bname}_{col}'] = bandpower_welch(sig, fs, brange) if len(sig)>fs else np.nan
                        else:
                             feat[f'{bname}_{col}'] = np.nan # Handle case where column might be missing in segment

        rows.append({**trial.to_dict(), **feat})
    return pd.DataFrame(rows)
```

```python
eeg_bandpower_df = compute_bandpower_per_trial(df_eeg, df_psy, eeg_band_cols, fs=2.3)
```

## Facial features
Detect emotion prob columns and AU columns from your TIVA dataframe

```python
# lower-case column names for detection
tiva_columns_lower = [c.lower() for c in df_tiva.columns]

# candidate emotion names
emotion_candidates = ['joy','anger','fear','sadness','disgust','surprise','engagement','valence','arousal','happy','neutral','attention','sentimentality','confusion','contempt']
emotion_cols = [c for c in df_tiva.columns if c.lower() in emotion_candidates]

# action unit columns (you have no 'AU_' prefix but you have AU-like names)
au_cols = [c for c in df_tiva.columns if c not in ['UnixTime','Row','QuestionKey','Timestamp','SampleNumber','Student_ID','File_Type'] and c not in emotion_cols]

print("emotion_cols:", emotion_cols)
print("au_cols:", au_cols)
```

Aggregate per trial (mean, std, max, occurrence counts):

```python
def aggregate_tiva_per_trial(df_psy, df_tiva, emotion_cols, au_cols,
                             time_col='UnixTime', student_col='Student_ID'):
    rows=[]
    by_student = {sid:g for sid,g in df_tiva.groupby(student_col)}
    for _, trial in tqdm(df_psy.iterrows(), total=len(df_psy), desc='TIVA trial agg'):
        sid=trial[student_col]; s=trial['routineStart']; e=trial['routineEnd']
        seg = by_student.get(sid)
        feat={}
        if seg is None:
            seg = pd.DataFrame()
        else:
            seg = seg[(seg[time_col]>=s) & (seg[time_col]<=e)]
        # emotion probs
        for c in emotion_cols:
            feat[f'{c}_mean'] = seg[c].mean() if (c in seg.columns and not seg.empty) else np.nan
            feat[f'{c}_std']  = seg[c].std() if (c in seg.columns and not seg.empty) else np.nan
            feat[f'{c}_max']  = seg[c].max() if (c in seg.columns and not seg.empty) else np.nan
        # AU stats
        for au in au_cols:
            if au in seg.columns and not seg.empty:
                feat[f'{au}_mean'] = seg[au].mean()
                feat[f'{au}_occurrence'] = (seg[au] > 0.1).sum()  # threshold for occurrence
            else:
                feat[f'{au}_mean'] = np.nan
                feat[f'{au}_occurrence'] = 0
        rows.append({**trial.to_dict(), **feat})
    return pd.DataFrame(rows)
```

```python
tiva_trial_feats = aggregate_tiva_per_trial(df_psy, df_tiva, emotion_cols, au_cols, time_col='UnixTime')
```

## Combine EEG + TIVA features into **df_features**

```python
df_features = df_psy.copy().reset_index(drop=True)

if 'eeg_trial_feats' in globals():
    extra = [c for c in eeg_trial_feats.columns if c not in df_psy.columns]
    df_features = pd.concat([df_features, eeg_trial_feats[extra].reset_index(drop=True)], axis=1)

if 'tiva_trial_feats' in globals():
    extra = [c for c in tiva_trial_feats.columns if c not in df_psy.columns]
    df_features = pd.concat([df_features, tiva_trial_feats[extra].reset_index(drop=True)], axis=1)
```
